```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import display, Markdown, Latex
from scipy.optimize import curve_fit
import matplotlib.animation as ani
import matplotlib.ticker as ticker
%matplotlib inline

timestep=2 #ns, must be the same for the two files
thickness_of_sample=101300 #micrometers
maxTempReached=1000 #K, is written in the end line of file .mlt
filename1='SIO2.20nm.td'
filename2='SIO2.40nm.td'
filename3='SIO2.60nm.td'
filename4='SIO2.80nm.td'
```

```python
data1 = np.loadtxt(filename1)
data2 = np.loadtxt(filename2)
data3 = np.loadtxt(filename3)
data4 = np.loadtxt(filename4)
x1 = data1[:,0]
X=data1[:,:]
for i in range(1,len(X[0,:])):
    y1 = data1[:,i]
    y2 = data2[:,i]
    y3 = data3[:,i]
    y4 = data4[:,i]
    t=(i)*timestep
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(x1, y1, c='navy', label='20 nm')
    ax.plot(x1, y2, c='crimson',label='40 nm')
    ax.plot(x1, y3, c='k',label='60 nm')
    ax.plot(x1, y4, c='orange', label='80 nm')
    ax.set_ylim(250,maxTempReached)
    ax.set_xlim(-100,4000)
    ax.set_title('timestep %i ns' %t,fontsize=25)
    ax.set_xlabel('depth ($\AA$)', fontsize=18)
    ax.set_ylabel('Temperature (K)', fontsize=18)
    plt.legend()
    plt.show()
```

```python
import matplotlib.gridspec as gridspec

data1 = np.loadtxt(filename1)
data2 = np.loadtxt(filename2)
data3 = np.loadtxt(filename3)
data4 = np.loadtxt(filename4)
x1 = data1[:,0]
X=data1[:,:]
n_cols=int(len(X[0,:])/5)#with different timestep check how many rows you desire
fig, ax = plt.subplots(5, n_cols+1, figsize=(70,35))#with different timestep check how many rows you desire
ax = ax.ravel()
#fig.suptitle('Temperature (K) vs depth for energy impulse %i $mJ/cm^2$' %energy_impulse,fontsize=70)
for i in range(len(X[0,:])-1):
    y1 = data1[:,i+1]
    y2 = data2[:,i+1]
    y3 = data3[:,i+1]
    y4 = data4[:,i+1]
    t=(i)*timestep
    ax[i].plot(x1, y1, c='navy', label='20 nm')
    ax[i].plot(x1, y2, c='crimson',label='40 nm')
    ax[i].plot(x1, y3, c='k',label='60 nm')
    ax[i].plot(x1, y4, c='orange', label='80 nm')
    ax[i].set_ylim(250,maxTempReached)
    ax[i].set_xlim(-100,4000)
    ax[i].set_title('timestep %i ns' %t,fontsize=25)
    ax[i].set_xlabel('depth ($\AA$)', fontsize=25)
    ax[i].set_ylabel('Temperature (K)', fontsize=25)
ax[0].legend(loc='upper left', fontsize=25)
fig.tight_layout()
plt.savefig('plot_Tvsdepth_50mJ')
```
