### KrF pulse Padova

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import display, Markdown, Latex
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
data=np.loadtxt('KrFpadova.txt')
ns=data[:,0]
au=data[:,1]
plt.figure(figsize=(10,6))
plt.plot(ns, au, color='red', label='pulse')
plt.title('KrF pulse laser',fontsize=18)
plt.xlabel('Time (ns)',fontsize=18)
plt.ylabel('a.u.',fontsize=18)
plt.legend(fontsize=16)
plt.show()
```
